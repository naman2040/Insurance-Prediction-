<hr>
<h2><center><b>Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<hr>
<h2><center><b>Loading Dataset

```python
df=pd.read_csv("Medical_Cost_Prediction.csv")
df
```

<hr>
<h2><center><b>Data Preprocessing

```python
df["sex"]=df["sex"].replace({"female":0,"male":1})
df["smoker"]=df["smoker"].replace({"no":0,"yes":1})
```

```python
x=df.iloc[:,:-1]
y=df["charges"].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
col_trans=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[5])],remainder='passthrough')
x=np.array(col_trans.fit_transform(x))
```

```python
y=y.reshape((-1,1))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

<hr>
<h2><center><b>Machine Learning Models</h2></center<br>
<h2>Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print("Result of Multiple Linear Regression Model on data is : ",r2_score(y_test,y_pred))
```

# SVR Model

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_std=ss.fit_transform(x_train)
x_test_std=ss.transform(x_test)

from sklearn.svm import SVR
for i in ["poly","sigmoid","rbf","linear"]:
    reg_svr=SVR(kernel=i)
    reg_svr.fit(x_std,y_train.ravel())
    y_pred=reg_svr.predict(x_test_std)
    print("####",i,"\nResult of SVR Model on data is : ",r2_score(y_test,y_pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
reg_tree=DecisionTreeRegressor(random_state=1)
reg_tree.fit(x_std,y_train)
y_pred_d_tree=reg_tree.predict(x_test_std)
print("Result of Decision tree Model on data is : ",r2_score(y_test,y_pred_d_tree))
```

# Random Forest Reg

```python
from sklearn.ensemble import RandomForestRegressor
reg_ran=RandomForestRegressor(n_estimators=300,random_state=0)
reg_ran.fit(x_std,y_train.ravel())
y_pred_r_forest=reg_ran.predict(x_test_std)
print("Result of Random Forest Model on data is : ",r2_score(y_test,y_pred_r_forest))
```


<h4>Thus it can be seen that "Random Forest Regression Model" has the best fit to out dataset :</h4>
<b>i.e R square value = 0.857</b>

<h2><b>Artificial Neural Network 

```python
import tensorflow as tf
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=9,activation="relu"))
ann.add(tf.keras.layers.Dense(units=9,activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=1))
```

```python
ann.compile(optimizer="adam",loss="mean_squared_error")
ann.fit(x_std,y_train,epochs=500)
```

```python
y_pred=ann.predict(x_test_std)
print(r2_score(y_test,y_pred))
```
